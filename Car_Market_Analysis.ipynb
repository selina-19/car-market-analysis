## Feature Engineering


## Mileage Condition

```python
bins = [0, 30000, 70000, float('inf')]
labels = ['Low', 'Medium', 'High']
df['mileage_category'] = pd.cut(df['mileage'], bins=bins, labels=labels, right=False)
print(df[['mileage', 'mileage_category']].head())
```

```python
mileage_category_summary = df.groupby('mileage_category')['price'].describe()
print(mileage_category_summary)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='mileage_category', y='mileage', data=df)
plt.title('Boxplot of Mileage by Mileage Category')
plt.ylabel('Mileage')
plt.xlabel('Mileage Category')
plt.show()
```

```python
df['mileage_bucket_encoded'] = df['mileage_category'].map({'Low': 1, 'Medium': 2, 'High': 3})
```

```python
correlation = df['mileage_bucket_encoded'].corr(df['price'])
print(f"Correlation between mileage_category and price: {correlation}")
```

```python
df['vehicle_condition_encoded'] = df['vehicle_condition'].map({'NEW': 1, 'USED': 0})
```

```python
correlation = df['vehicle_condition_encoded'].corr(df['mileage_bucket_encoded'])
print(f"Correlation between vehicle_condition and price: {correlation}")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='mileage_category', y='price', data=df)
plt.title('Box Plot of Mileage Category vs. Price')
plt.xlabel('Mileage Category')
plt.ylabel('Price')
plt.show()
```

Mileage Condition

```python
df['mileage_condition'] = df['mileage_category'].astype(str) + '_' + df['vehicle_condition'].astype(str)
print(df['mileage_condition'].value_counts())
```

```python
plt.clf()
```

```python
top_n = 150
top_categories = df['brand_body'].value_counts().nlargest(top_n).index
selected_categories = []
suffixes_to_limit = ['suv', 'mpv']
for suffix in suffixes_to_limit:
    categories_with_suffix = [cat for cat in top_categories if cat.endswith(suffix)]
    selected_categories.extend(categories_with_suffix[:2])

remaining_categories = [cat for cat in top_categories if cat not in selected_categories]
selected_categories.extend(remaining_categories[:5])

df_filtered = df[df['brand_body'].isin(selected_categories)]

plt.figure(figsize=(12, 8))
sns.boxplot(x='brand_body', y='price', hue='crossover_car_and_van', data=df_filtered)
plt.xticks(rotation=45)
plt.title('Box Plot of Price by Brand_body')
plt.xlabel('Brand_body')
plt.ylabel('Price')
plt.show()
```
